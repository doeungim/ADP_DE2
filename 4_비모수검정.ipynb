### âš ï¸ ì •ê·œì„±/ë“±ë¶„ì‚°ì„± ì¡°ê±´ì— ë”°ë¥¸ ëŒ€ì•ˆ ê²€ì •

| ì •ê·œì„± | ë“±ë¶„ì‚°ì„± | ì ìš© ê²€ì •                       | Python í•¨ìˆ˜ ì˜ˆì‹œ                                   | ì‚¬í›„ ê²€ì • í•¨ìˆ˜ ì˜ˆì‹œ                             |
|--------|----------|-------------------------------|---------------------------------------------------|------------------------------------------------|
| âœ…     | âŒ       | Welch ANOVA                   | `scipy.stats.ttest_ind(..., equal_var=False)`     | `pingouin.pairwise_gameshowell`                |
| âŒ     | âœ…       | Kruskal-Wallis ê²€ì •           | `scipy.stats.kruskal`                             | `scikit_posthocs.posthoc_dunn`                 |
| âŒ     | âŒ       | Kruskal-Wallis ë˜ëŠ” ë¶€íŠ¸ìŠ¤íŠ¸ë© | `scipy.stats.kruskal`, `sklearn.utils.resample`   | `scikit_posthocs.posthoc_dunn`, `bootstrapped.bootstrap` |


````python
# ì˜ˆì œ ë°ì´í„°ë¥¼ ìƒì„±í•˜ê³  ì£¼ìš” ë¹„ëª¨ìˆ˜ ê²€ì •ì„ ì§ì ‘ ì ìš©í•´ë³´ëŠ” ì½”ë“œ
import numpy as np
import pandas as pd
from scipy.stats import mannwhitneyu, wilcoxon, kruskal, friedmanchisquare, chi2_contingency, spearmanr, kendalltau

# -----------------------------
# âœ… ì˜ˆì œ 1: Mann-Whitney U test (ë‘ ë…ë¦½ ì§‘ë‹¨ ë¹„êµ)
group_A = [82, 85, 88, 90, 91]
group_B = [78, 80, 84, 87, 89]

mw_stat, mw_p = mannwhitneyu(group_A, group_B)

# -----------------------------
# âœ… ì˜ˆì œ 2: Wilcoxon Signed-Rank test (ë‘ ê´€ë ¨ ì§‘ë‹¨ ë¹„êµ)
before = [120, 130, 115, 140, 125]
after = [122, 128, 117, 142, 126]

wilcoxon_stat, wilcoxon_p = wilcoxon(before, after)

# -----------------------------
# âœ… ì˜ˆì œ 3: Kruskal-Wallis H test (3ê°œ ë…ë¦½ ì§‘ë‹¨ ë¹„êµ)
group_X = [5.2, 5.5, 5.1, 5.3]
group_Y = [6.0, 6.1, 6.2, 6.1]
group_Z = [5.7, 5.8, 5.9, 6.0]

kruskal_stat, kruskal_p = kruskal(group_X, group_Y, group_Z)

# -----------------------------
# âœ… ì˜ˆì œ 4: Friedman Test (3ê°œ ì´ìƒ ëŒ€ì‘ ì§‘ë‹¨)
cond1 = [7, 8, 6, 9]
cond2 = [6, 7, 5, 8]
cond3 = [8, 9, 7, 10]

friedman_stat, friedman_p = friedmanchisquare(cond1, cond2, cond3)

# -----------------------------
# âœ… ì˜ˆì œ 5: Chi-Squared Test (ë¶„í• í‘œ)
table = [[10, 20], [20, 30]]
chi2_stat, chi2_p, _, _ = chi2_contingency(table)

# -----------------------------
# âœ… ì˜ˆì œ 6: Spearman & Kendall ìƒê´€
x = [10, 20, 30, 40, 50]
y = [12, 18, 33, 39, 52]

spearman_corr, spearman_p = spearmanr(x, y)
kendall_corr, kendall_p = kendalltau(x, y)

# ê²°ê³¼ ì •ë¦¬
results = pd.DataFrame({
    'Test': [
        'Mann-Whitney U', 'Wilcoxon Signed-Rank', 'Kruskal-Wallis H',
        'Friedman Test', 'Chi-Squared Test', 'Spearman Correlation', 'Kendall Tau Correlation'
    ],
    'Statistic': [
        mw_stat, wilcoxon_stat, kruskal_stat,
        friedman_stat, chi2_stat, spearman_corr, kendall_corr
    ],
    'p-value': [
        mw_p, wilcoxon_p, kruskal_p,
        friedman_p, chi2_p, spearman_p, kendall_p
    ]
})

results
````

- (2) ê·€ë¬´ê°€ì„¤ ë° ìœ ì˜ìˆ˜ì¤€ 0.05 í•˜ì—ì„œì˜ ê²°ë¡ ì„ ì œì‹œí•˜ì‹œì˜¤.

````python
from statsmodels.sandbox.stats.runs import runstest_2samp
sensor_A = [22, 24, 23, 21, 22]
sensor_B = [28, 29, 30, 31, 29]

median_A = np.median(sensor_A)
median_B = np.median(sensor_B)
# 1,0ìœ¼ë¡œ ì´ì§„í™”
A = np.array([1.0 if k >= median_A else 0 for k in sensor_A])
B = np.array([1.0 if k >= median_B else 0 for k in sensor_B])

stat, p_val = runstest_2samp(A, B)
print(p_val)
````

### ğŸ”¹ ë¶€í˜¸ê²€ì •
1ï¸âƒ£ ì¼í‘œë³¸
- ì¤‘ì•™ê°’ì´ ë‹¤ë¥¸ì§€, ê´€ì¸¡ê°’ì„ +, -, 0ìœ¼ë¡œ ë³€í™˜í•˜ì—¬ +ì˜ ê°œìˆ˜ í˜¹ì€ -ì˜ ê°œìˆ˜ë¡œ ê²€ì •
#####  âœ… í‘œë³¸ ìˆ˜ (n)	ì‚¬ìš©í•˜ëŠ” ë¶„í¬
* n â‰¤ 25	: ì •í™•í•œ ì´í•­ë¶„í¬ ì‚¬ìš© (B(n, 0.5))
* 25 < n â‰¤ 100	: ì´í•­ë¶„í¬ ê·¼ì‚¬ ë˜ëŠ” ì •ê·œ ê·¼ì‚¬ ê°€ëŠ¥
* n > 100 :	ì •ê·œë¶„í¬ ê·¼ì‚¬ ì‚¬ìš© ê°€ëŠ¥ (Z í†µê³„ëŸ‰ ì‚¬ìš©)

````python
data = np.array([203,204,197,195,201,205,198,199,194,207]) #ë°ì´í„°
M0 = 200  # ì¤‘ìœ„ìˆ˜

# H0 : ë°ì´í„°ì˜ ì¤‘ì•™ê°’ì€ 200ê³¼ ì°¨ì´ê°€ ì—†ë‹¤.
# H1 : ë°ì´í„°ì˜ ì¤‘ì•™ê°’ì€ 200ê³¼ ì°¨ì´ê°€ ìˆë‹¤.

from scipy.stats import binomtest
# data ì™€ M0ì˜ ì°¨ì´ì— SINGìœ¼ë¡œ ì–‘ìˆ˜ë©´ 1, ìŒìˆ˜ë©´ -1
signs = np.sign(data - M0)
positive_signs = np.sum(signs > 0)
n = np.sum(signs != 0)

# Binomial test
p_value = binomtest(positive_signs, n=n, p=0.5, alternative='two-sided')
print(f'One-Sample Sign Test p-value: {p_value}')
````

````python
# ì˜ˆì œ2) ë¬¸ì œ
#ì•„ë˜ ë°ì´í„°ì˜ ì¤‘ì•™ê°’ì´ 130ì´ë¼ê³  í•  ìˆ˜ ìˆëŠ”ì§€ ê²€ì •í•˜ë¼
d = np.array([133,145,155,162,143,133,166])
m = 130

sign = np.sign(d-m)
positive_sign = np.sum(sign > 0)
n = np.sum(sign != 0)

p_val = binomtest(positive_sign, n = n, p = 0.5, alternative = 'two-sided')
print(p_val)
````

2. ìœ ì˜ìˆ˜ì¤€ 0.05ì—ì„œ ê·€ë¬´ê°€ì„¤ì„ ê¸°ê°í•  ìˆ˜ ìˆëŠ”ì§€ íŒë‹¨í•˜ì‹œì˜¤.
3. ì •ê·œì„± ë§Œì¡± ì‹œ ì–´ë–¤ ê²€ì •ì„ ì‚¬ìš©í•  ìˆ˜ ìˆëŠ”ì§€ë„ ëª…ì‹œí•˜ì‹œì˜¤.

````python
from scipy.stats import binomtest
old = np.array([12, 14, 15, 13, 16, 12, 13, 14, 15, 17])
new = np.array([11, 13, 13, 12, 15, 11, 11, 13, 14, 15])

# ì „í›„ ì°¨ì´
diff = old - new
# ë¶€í˜¸ ê³„ì‚°
pos = np.sum(diff > 0 )
neg = np.sum(diff < 0)
n = pos + neg       # 0ì€ ì œì™¸

res = binomtest(pos, n = n , p = 0.5, alternative= 'two-sided')
print(res.pvalue)
````

### ğŸ”¹ ìœŒì½•ìŠ¨ ë¶€í˜¸ ìˆœìœ„ ê²€ì • (Singed-Rank Test)
### ğŸ“Š ë¶€í˜¸ê²€ì • vs ìœŒì½•ìŠ¨ ë¶€í˜¸ìˆœìœ„ê²€ì • ë¹„êµí‘œ
$$
\begin{array}{|c|c|c|}
\hline
\textbf{í•­ëª©} & \textbf{ë¶€í˜¸ê²€ì • (Sign Test)} & \textbf{ìœŒì½•ìŠ¨ ë¶€í˜¸ìˆœìœ„ê²€ì • (Wilcoxon Signed-Rank Test)} \\
\hline
ê²€ì • ëª©ì  & ëŒ€ì‘í‘œë³¸ ê°„ ì¤‘ì•™ê°’ ì°¨ì´ ê²€ì • & ëŒ€ì‘í‘œë³¸ ê°„ ì¤‘ì•™ê°’ ì°¨ì´ ê²€ì • (ë” ë¯¼ê°) \\
\hline
ì‚¬ìš© ì •ë³´ & ì°¨ì´ì˜ ë¶€í˜¸ë§Œ ì‚¬ìš© (+, âˆ’) & ì°¨ì´ì˜ ì ˆëŒ“ê°’ê³¼ ìˆœìœ„ê¹Œì§€ ì‚¬ìš© \\
\hline
í†µê³„ëŸ‰ ê³„ì‚° ë°©ì‹ & + ë˜ëŠ” âˆ’ ë¶€í˜¸ì˜ ê°œìˆ˜ â†’ ì´í•­ê²€ì • & ì ˆëŒ“ê°’ì„ ìˆœìœ„í™” í›„ ë¶€í˜¸ì™€ ê³±í•˜ì—¬ í•©ì‚° \\
\hline
ê²€ì •ë ¥ (íŒŒì›Œ) & ë‚®ìŒ (ë³´ìˆ˜ì ) & ë†’ìŒ (ë” ë¯¼ê°í•˜ê²Œ ì°¨ì´ë¥¼ íƒì§€í•¨) \\
\hline
ì°¨ì´ì˜ í¬ê¸° ê³ ë ¤ ì—¬ë¶€ & ê³ ë ¤í•˜ì§€ ì•ŠìŒ & ê³ ë ¤í•¨ \\
\hline
ì œì™¸ë˜ëŠ” ê´€ì¸¡ì¹˜ & ì°¨ì´ê°€ 0ì¸ ìŒ & ì°¨ì´ê°€ 0ì¸ ìŒ \\
\hline
Python í•¨ìˆ˜ ì˜ˆì‹œ & \texttt{scipy.stats.binomtest} & \texttt{scipy.stats.wilcoxon} \\
\hline
ë¶„í¬ ê°€ì • & ì—†ìŒ & ëŒ€ì¹­ì„± ê°€ì • í•„ìš” (ì •ê·œì„±ì€ í•„ìš” ì—†ìŒ) \\
\hline
ì í•©í•œ ìƒí™© ì˜ˆì‹œ & ì¢‹ì•„ì§/ë‚˜ë¹ ì§ ë“± ë°©í–¥ë§Œ ì¤‘ìš”í•  ë•Œ & ì°¨ì´ì˜ í¬ê¸°ê¹Œì§€ ì˜ë¯¸ ìˆëŠ” ê²½ìš° \\
\hline
\end{array}
$$


````python
from scipy.stats import wilcoxon

# H0 : ë°ì´í„° ì¤‘ìœ„ìˆ˜ = 200
# H1 : ë°ì´í„° ì¤‘ìœ„ìˆ˜ != 200
data = np.array([203,204,197,195,201,205,198,199,194,207]) #ë°ì´í„°
M0 = 200 #ì¤‘ì•™ê°’

stat, p_value = wilcoxon(M0 -data)
print(f'Wilcoxon statistic: {stat}, p-value: {p_value}')

stats, p = wilcoxon([M0 for i in range(len(data))], data)
print(stats, p)
````

### ğŸ”¹ ìœŒì½•ìŠ¨ ë¶€í˜¸ ìˆœìœ„í•© (Rank-Sum Test)

### ğŸ“Š ìœŒì½•ìŠ¨ ê²€ì • ë° ë§Œ-íœ˜íŠ¸ë‹ˆ U ê²€ì • ë¹„êµ
$$
\begin{array}{|c|c|c|c|}
\hline
\textbf{í•­ëª©} & \textbf{ìœŒì½•ìŠ¨ ë¶€í˜¸ìˆœìœ„ê²€ì •} & \textbf{ìœŒì½•ìŠ¨ ìˆœìœ„í•©ê²€ì •} & \textbf{ë§Œ-íœ˜íŠ¸ë‹ˆ U ê²€ì •} \\
\hline
ê²€ì • ëª©ì  & ë™ì¼ ì§‘ë‹¨ì˜ ì „/í›„ ë³€í™” ê²€ì • (ì¤‘ì•™ê°’ ì°¨ì´) & ì„œë¡œ ë‹¤ë¥¸ ë‘ ì§‘ë‹¨ì˜ ì¤‘ì•™ê°’ ì°¨ì´ ê²€ì • & ì„œë¡œ ë‹¤ë¥¸ ë‘ ì§‘ë‹¨ì˜ ì¤‘ì•™ê°’ ì°¨ì´ ê²€ì • \\
\hline
ë‹¤ë¥¸ ì´ë¦„ & ëŒ€ì‘í‘œë³¸ ìœŒì½•ìŠ¨ & ë¹„ëŒ€ì‘ ìœŒì½•ìŠ¨, ìˆœìœ„í•© ê²€ì • & ìœŒì½•ìŠ¨ ìˆœìœ„í•©ê²€ì •ê³¼ ë™ì¼ ê°œë… \\
\hline
ì ìš© ë°ì´í„° & ê°™ì€ ëŒ€ìƒì˜ ë‘ ì¡°ê±´ (paired) & ë…ë¦½ëœ ë‘ ê·¸ë£¹ (independent) & ë…ë¦½ëœ ë‘ ê·¸ë£¹ (independent) \\
\hline
ë°ì´í„° ëŒ€ì‘ì„± & ëŒ€ì‘ (paired samples) & ë¹„ëŒ€ì‘ (independent samples) & ë¹„ëŒ€ì‘ (independent samples) \\
\hline
Python í•¨ìˆ˜ & \texttt{scipy.stats.wilcoxon} & \texttt{scipy.stats.ranksums} & \texttt{scipy.stats.mannwhitneyu} \\
\hline
ë¶„í¬ ê°€ì • & ì •ê·œì„± í•„ìš” ì—†ìŒ, ì°¨ì´ì˜ ëŒ€ì¹­ì„± í•„ìš” & ì •ê·œì„± í•„ìš” ì—†ìŒ & ì •ê·œì„± í•„ìš” ì—†ìŒ \\
\hline
ì˜ˆì‹œ ìƒí™© & ë³µìš© ì „í›„ í˜ˆì•• ë³€í™” & ì‹¤í—˜êµ° vs ëŒ€ì¡°êµ° ë¹„êµ & ì‹¤í—˜êµ° vs ëŒ€ì¡°êµ° ë¹„êµ \\
\hline
\end{array}
$$


````python
# ì˜ˆì œ1) íŒ€ ì˜ì—…ë³„ ì„±ì ì˜ ì°¨ì´ê°€ ìˆëŠ”ì§€ ê²€ì •
# H0 : A,B ì¤‘ì•™ê°’ì˜ ì°¨ì´ëŠ” 0 ì´ë‹¤.
# H1 : A,B ì¤‘ì•™ê°’ì˜ ì°¨ì´ëŠ” 0 ì´ ì•„ë‹ˆ.

data1 = [87,75,65,95,90,81,93]
data2 = [57,85,90,83,87,71]

print("[ë¼ì´ë¸ŒëŸ¬ë¦¬ ê²€ì •]")
from scipy.stats import ranksums, mannwhitneyu
zstats, p = ranksums(data1, data2) #default: alternative='two-sided'
print(" ìˆœìœ„í•© ì •ê·œê·¼ì‚¬ ê²€ì •í†µê³„ëŸ‰: {:.3f}, p-value: {:.3f}".format(zstats, p))


mstat, p = mannwhitneyu(data1, data2, alternative='two-sided')
print(" ë§Œìœ„íŠ¸ë‹ˆ ê²€ì •í†µê³„ëŸ‰: {:.3f}, p-value: {:.3f}".format(mstat, p))
````
